# Analyzing hubness connection

#### For each attr. we have a (x,y) coordinate file, where x is the hubness of the query entity and y the number of hits obtained.

```python
import matplotlib.pyplot as plt
import matplotlib.axes as axes
import seaborn as sns
import numpy as np
from collections import defaultdict
```

```python
def get_x_y(path):
    x = []
    y = []
    with open(path, 'r') as f:
        for line in f.readlines():
            tokens = line.split(",")
            q_hubness = float(tokens[0])
            hits = int(tokens[1])
            x.append(q_hubness)
            y.append(hits)
    return x, y
```

## Mit Id

```python
# Mit Id
path = "mitId_qh_coordinates.txt"
x, y= get_x_y(path)
```

```python
plt.plot(x, y, '.')
plt.xlabel('Hubness')
plt.ylabel('Hits')
plt.title('Mit Id')
plt.show()
```

```python
x = np.asarray(x)
y = np.asarray(y)
sns.regplot(x, y)
```

```python
plt.hexbin(x, y, gridsize=50, cmap='inferno', bins='log')
```
